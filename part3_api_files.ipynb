```python
# Task 1 — File Read & Write Basics
# Part B — Reading the file
try:
    with open(FILENAME, "r", encoding="utf-8") as file:
        lines = file.readlines()

    print("\n--- File Contents ---")
    for index, line in enumerate(lines, start=1):
        print(f"{index}. {line.strip()}")

    print(f"\nTotal number of lines: {len(lines)}")

    # Keyword search
    keyword = input("\nEnter a keyword to search: ").strip().lower()
    matches = [
        line.strip() for line in lines if keyword in line.lower()
    ]

    if matches:
        print("\nLines containing the keyword:")
        for match in matches:
            print(match)
    else:
        print("No lines found containing the given keyword.")

except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print(f"Error while reading file: {e}")
```

```python
#Task 2 — API Integration
import requests
from pprint import pprint

BASE_URL = "https://dummyjson.com/products"


# Step 1 — Fetch & Display Products

def fetch_products(limit=20):
    try:
        response = requests.get(f"{BASE_URL}?limit={limit}", timeout=10)
        response.raise_for_status()
        return response.json()["products"]
    except requests.RequestException as e:
        print(f"Error fetching products: {e}")
        return []


def print_product_table(products):
    print("\nID  | Title                          | Category      | Price     | Rating")
    print("----|--------------------------------|---------------|-----------|-------")

    for p in products:
        print(
            f"{p['id']:<4}| "
            f"{p['title'][:30]:<30} | "
            f"{p['category']:<13} | "
            f"${p['price']:<9} | "
            f"{p['rating']}"
        )


# Step 2 — Filter & Sort

def filter_and_sort(products):
    filtered = [p for p in products if p["rating"] >= 4.5]
    filtered.sort(key=lambda x: x["price"], reverse=True)
    return filtered



# Step 3 — Search by Category

def fetch_category(category):
    try:
        response = requests.get(
            f"{BASE_URL}/category/{category}", timeout=10
        )
        response.raise_for_status()
        return response.json()["products"]
    except requests.RequestException as e:
        print(f"Error fetching category '{category}': {e}")
        return []



# Step 4 — POST Request (Simulated)

def add_custom_product():
    payload = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    try:
        response = requests.post(
            f"{BASE_URL}/add",
            json=payload,
            timeout=10
        )
        response.raise_for_status()
        return response.json()
    except requests.RequestException as e:
        print(f"Error creating product: {e}")
        return None

# Main

if __name__ == "__main__":

    # Step 1
    products = fetch_products()
    print_product_table(products)

    # Step 2
    print("\n\n--- Products with Rating ≥ 4.5 (Sorted by Price Desc) ---")
    high_rated = filter_and_sort(products)
    print_product_table(high_rated)

    # Step 3
    print("\n\n--- Laptops Category ---")
    laptops = fetch_category("laptops")
    for laptop in laptops:
        print(f"{laptop['title']} — ${laptop['price']}")

    # Step 4
    print("\n\n--- POST Request Response ---")
    created_product = add_custom_product()
    if created_product:
        pprint(created_product)
```

```python
# Task 3 — Exception Handling
# Part A — Guarded Calculator
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


# Tests
print(safe_divide(10, 2))      # Expected: 5.0
print(safe_divide(10, 0))      # Expected: Error: Cannot divide by zero
print(safe_divide("ten", 2))   # Expected: Error: Invalid input types
```

```python
# Task 3 — Exception Handling
# Part B — Guarded File Reader

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")


# Tests
print("\n--- Reading existing file ---")
content = read_file_safe("python_notes.txt")
if content:
    print(content)

print("\n--- Reading missing file ---")
read_file_safe("ghost_file.txt")
```

```python
# Task 3 — Exception Handling
# Part C — Robust API Calls

import requests

BASE_URL = "https://dummyjson.com/products"

def safe_get(url):
    try:
        response = requests.get(url, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None


def safe_post(url, payload):
    try:
        response = requests.post(url, json=payload, timeout=5)
        return response
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None
```

```python
# Task 3 — Exception Handling
# Part D — Input Validation Loop

while True:
    user_input = input(
        "\nEnter a product ID to look up (1–100), or 'quit' to exit: "
    ).strip().lower()

    if user_input == "quit":
        print("Exiting product lookup.")
        break

    if not user_input.isdigit():
        print("Invalid input. Please enter a number between 1 and 100.")
        continue

    product_id = int(user_input)

    if not (1 <= product_id <= 100):
        print("Invalid ID. Please enter a number between 1 and 100.")
        continue

    response = safe_get(f"{BASE_URL}/{product_id}")

    if response is None:
        continue

    if response.status_code == 404:
        print("Product not found.")
    elif response.status_code == 200:
        product = response.json()
        print(f"Product: {product['title']} | Price: ${product['price']}")
    else:
        print(f"Unexpected server response: {response.status_code}")
```

```python
# Task 4 — Logging to File
from datetime import datetime
import requests

LOG_FILE = "error_log.txt"

def log_error(context, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {context}: {error_type} — {message}\n"
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(log_entry)
```
